```python
import pandas as pd
import seaborn as sns
import gdown
url='https://drive.google.com/file/d/14qB9jN9e-dWQtNiZeS8rHdGwMAE54PHA/view?usp=drive_link'
output='dataset.csv'
gdown.download(url,output,quiet=False,fuzzy=True)

df=pd.read_csv(output)
df
```

```python
df.describe()
```

```python
sns.heatmap(df.isnull())
df.isnull().sum().sum()
df.describe().T
#this is used to deal with nan balues
'''
All features are encoded as -1 (phishing indicator),
0 (uncertain), or 1 (legitimate indicator) with no missing values.
Features with mean near -1 such as Prefix_Suffix (-0.73) and
URLURL_Length (-0.63) suggest these signals lean phishing across the
dataset. Features with high standard deviation (>0.8) such as
Request_URL and having_Sub_Domain show strong variation
and are likely the most discriminative for classification.
'''
```

```python
df.columns=df.columns.str.strip()
df
```

```python
from matplotlib import pyplot as plt
feature_cols = [c for c in df.columns if c not in ['index', 'Result']]

fig, axes = plt.subplots(6, 5, figsize=(18, 16))
axes = axes.flatten()

for i, col in enumerate(feature_cols):
    leg = df[df['Result'] == 1][col].value_counts().sort_index()
    phish = df[df['Result'] == -1][col].value_counts().sort_index()

    x = [-1, 0, 1]
    leg_counts   = [leg.get(v, 0) for v in x]
    phish_counts = [phish.get(v, 0) for v in x]

    width = 0.35
    axes[i].bar([-0.2, 0.8, 1.8], leg_counts,
                width=width, label='Legitimate', color='#2ecc71', alpha=0.8)
    axes[i].bar([0.2, 1.2, 2.2], phish_counts,
                width=width, label='Phishing', color='#e74c3c', alpha=0.8)

    axes[i].set_title(col, fontsize=8, pad=4)
    axes[i].set_xticks([0, 1, 2])
    axes[i].set_xticklabels(['-1', '0', '1'], fontsize=7)
    axes[i].tick_params(axis='y', labelsize=7)
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)

axes[0].legend(fontsize=8)

for j in range(len(feature_cols), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Feature Value Distribution — Phishing vs Legitimate',
             fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```
