```python
from matplotlib import pyplot as plt
feature_cols = [c for c in df.columns if c not in ['index', 'Result']]

fig, axes = plt.subplots(6, 5, figsize=(18, 16))
axes = axes.flatten()

for i, col in enumerate(feature_cols):
    leg = df[df['Result'] == 1][col].value_counts().sort_index()
    phish = df[df['Result'] == -1][col].value_counts().sort_index()

    x = [-1, 0, 1]
    leg_counts   = [leg.get(v, 0) for v in x]
    phish_counts = [phish.get(v, 0) for v in x]

    width = 0.35
    axes[i].bar([-0.2, 0.8, 1.8], leg_counts,
                width=width, label='Legitimate', color='#2ecc71', alpha=0.8)
    axes[i].bar([0.2, 1.2, 2.2], phish_counts,
                width=width, label='Phishing', color='#e74c3c', alpha=0.8)

    axes[i].set_title(col, fontsize=8, pad=4)
    axes[i].set_xticks([0, 1, 2])
    axes[i].set_xticklabels(['-1', '0', '1'], fontsize=7)
    axes[i].tick_params(axis='y', labelsize=7)
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)

axes[0].legend(fontsize=8)

for j in range(len(feature_cols), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Feature Value Distribution — Phishing vs Legitimate',
             fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```

```python
X=df.drop(columns=['index','Result'])
y=df['Result']
print(X.shape)
print(y.shape)
```

```python
from sklearn.feature_selection import chi2,SelectKBest
from sklearn.preprocessing import MinMaxScaler
#chi square test requires non negative values
#our data has -1,0,1 so we scale to 0,,0.5,1
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
chi_scores,p_values=chi2(X_scaled,y)
chi_df=pd.DataFrame({
    'Feature':X.columns,
    'Chi2_score':chi_scores,
    'P_value':p_values
}).sort_values('Chi2_score',ascending=False)

print(chi_df.to_string(index=False))
```

```python
from sklearn.feature_selection import mutual_info_classif
mi_score=mutual_info_classif(X,y,random_state=42)
mi_df=pd.DataFrame({
    'Feature':X.columns,
    'MI_Score':mi_score
}).sort_values('MI_Score',ascending=False)
print(mi_df.to_string(index=False)
)
```

Both chi-square and mutual information consistently rank SSLfinal_State, URL_of_Anchor, Prefix_Suffix, web_traffic, having_Sub_Domain, and SFH as the top discriminative features. The agreement across two statistically independent methods with different mathematical assumptions confirms these features are genuinely informative. Features with MI score of exactly 0.000 (Shortining_Service, double_slash_redirecting, Favicon, on_mouseover, Iframe, Links_pointing_to_page) carry no measurable information about the label and will contribute noise rather than signal to linear models.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
print(f"Training set: {X_train.shape}")
print(f"test set: {X_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score

lr=LogisticRegression(max_iter=1000,random_state=42)
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)
y_prob_lr=lr.predict_proba(X_test)[:,1]

print("Logistic Regression")
print(confusion_matrix(y_test,y_pred_lr))
print(classification_report(y_test,y_pred_lr))
print(f"ROC: {roc_auc_score(y_test,y_prob_lr)}")
```

naive bayer has high bias but low variance


```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred_nb=nb.predict(X_test)
y_prob_nb=nb.predict_proba(X_test)[:,1]
print(confusion_matrix(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
print(f"ROC-AUC: {roc_auc_score(y_test, y_prob_nb)}")
```

Naive Bayes has 58% accuracy but 0.966 ROC-AUC.Because accuracy and ROC-AUC measure different things. Naive Bayes is flagging almost everything as phishing  hence 100% phishing recall but only 27% legitimate recall. It's being extremely aggressive. Terrible classifier in practice. But ROC-AUC measures the ranking quality across all thresholds  and Naive Bayes still ranks URLs reasonably well even if its default threshold is miscalibrated.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print(f"ROC-AUC: {roc_auc_score(y_test, y_prob_rf):.4f}")
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_xgb = le.fit_transform(y_train)
y_test_xgb = le.transform(y_test)
xgb = XGBClassifier(n_estimators=100, random_state=42,eval_metric='logloss', verbosity=0)
xgb.fit(X_train, y_train_xgb)
y_pred_xgb = le.inverse_transform(xgb.predict(X_test))
y_prob_xgb = xgb.predict_proba(X_test)[:,1]
print(confusion_matrix(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
print(f"ROC-AUC: {roc_auc_score(y_test, y_prob_xgb)}")
```

Naive Bayes assumes all features are conditionally independent given the class label. This assumption is violated in phishing URL data  SSLfinal_State, age_of_domain, and Page_Rank are correlated signals that co-occur in both phishing and legitimate URLs. Despite this, ROC-AUC remains 0.966 because the model's probability rankings are reasonable even when its threshold calibration fails. In a production system, Naive Bayes would be unusable due to its 72% false positive rate on legitimate URLs blocking legitimate sites nearly three times out of four.

```python
from sklearn.metrics import roc_curve, auc
fig, ax = plt.subplots(figsize=(8, 6))
models = {
    'Logistic Regression': y_prob_lr,
    'Naive Bayes':         y_prob_nb,
    'Random Forest':       y_prob_rf,
    'XGBoost':             y_prob_xgb,
}
colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']
for (name, probs), color in zip(models.items(), colors):
    fpr, tpr, _ = roc_curve(y_test, probs, pos_label=1)
    auc_score = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f'{name} (AUC={auc_score:.3f})',
            color=color, linewidth=2)

ax.plot([0,1],[0,1], 'k--', linewidth=1, label='Random baseline')
ax.set_xlabel('False Positive Rate (Legitimate URLs wrongly flagged)')
ax.set_ylabel('True Positive Rate (Phishing URLs caught)')
ax.set_title('ROC Curve Comparison — Phishing URL Detection')
ax.legend(loc='lower right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
